```python
def pipeline(pos, features=None):
    td = TransfersDataset(Path.cwd()/'../prepped/final-2.csv')
    td.filter_postion(pos)
    td.encode()
    td.drop()
    m = Model(data = td.data, features = features)
    display(m.data[m.features])
    m.train_test_split()
    m.scale()
    # m.tune_hp()
    m.train()
    m.inverse_scale()
    m.score()
    display(m.feature_importance(log_name=f'xgboost_{pos}'))
    m.plot_predictions()
    display(m.top_n_predictions(20))
```
